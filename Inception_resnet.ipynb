```python
import torch
import torch.nn as nn

# Define a custom PyTorch module for cosine similarity
class CosineSimilarity(nn.Module):
    def forward(self, x):
        vector1, vector2 = x

        # Calculate the dot product
        dot_product = torch.sum(vector1 * vector2, dim=-1)

        # Calculate the magnitudes of the vectors
        magnitude1 = torch.sqrt(torch.sum(vector1**2, dim=-1))
        magnitude2 = torch.sqrt(torch.sum(vector2**2, dim=-1))

        # Calculate cosine similarity
        similarity = dot_product / (magnitude1 * magnitude2)

        return similarity

# Assuming embeddings and embeddings1 are PyTorch tensors
cosine_similarity = CosineSimilarity()
cosine_similarity_Aaron_Peirsol = cosine_similarity([embeddings_test, embeddings_test1])
```
